```python
import requests
import wget
import os
from bs4 import BeautifulSoup, SoupStrainer
```

```python
#  connects Colab to Google Drive
from google.colab import drive
drive.mount('/content/drive/')
```

```python
# url with datasets
url_base = 'https://www.contextures.com/' 
url_endpoint = 'xlsampledata01.html'
response = requests.get(url_base + url_endpoint)

# creates soup with all links with 'a' tag
soup = BeautifulSoup(response.content, 'html.parser', parse_only=SoupStrainer('a'))

# folder where files will be downloaded
destination_path = '/content/drive/MyDrive/Python/Datasets/excel_samples'
```

```python
# valid file types
file_types = ['.rdata','.xls', '.csv','.doc','.pdf','zip']

# iterates through every data type
for file_type in file_types:
    
    file_counter = 0
    
    # waits for server response
    if (response.ok):
        print('Connecting to server...')
        
        # iterates through every links in soup
        for link in soup:
            if link.has_attr('href'):
              if file_type in link['href']:
                  full_path = url_base + link['href']
                  print('File found: ', full_path)
                  wget.download(full_path, out=destination_path)
                  file_counter+=1
        
        # displays file counter          
        print(file_counter, file_type, 'files found')
    
    else:
      print('Error: Unable to connect to server')

```
